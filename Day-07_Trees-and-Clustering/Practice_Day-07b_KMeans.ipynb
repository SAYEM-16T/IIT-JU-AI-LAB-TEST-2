## 1. Import + data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Live.csv')
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()   # Column1-4 sob NaN
```

## 2. Redundant/useless column drop
Column1-4 sob NaN. status_id, status_published prai unique ID -> useless.

```python
df.drop(['Column1','Column2','Column3','Column4'], axis=1, inplace=True)
df['status_type'].unique()   # ['video','photo','link','status']
```

```python
df.drop(['status_id','status_published'], axis=1, inplace=True)
df.head()
```

## 3. Categorical -> number + scaling
`LabelEncoder`: status_type -> 0/1/2/3. `MinMaxScaler`: sob feature 0-1 e (K-Means distance-based, tai scale joruri).

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['status_type'] = le.fit_transform(df['status_type'])
```

```python
from sklearn.preprocessing import MinMaxScaler
cols = df.columns
ms = MinMaxScaler()
df_scaled = ms.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=cols)
X = df_scaled.values
X[:5]
```

## 4. Elbow method — kotota cluster (K)?
K=1..10 er jonno inertia plot. Je jaygay line hothat komte thame (kohuni/elbow) = optimal K (~4).

```python
from sklearn.cluster import KMeans
clustering_score = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='random', random_state=42, n_init=10)
    kmeans.fit(X)
    clustering_score.append(kmeans.inertia_)  # inertia = within-cluster sum of sq distances

plt.figure(figsize=(10,6))
plt.plot(range(1, 11), clustering_score, marker='o')
plt.scatter(4, clustering_score[3], s=200, c='red', marker='*')
plt.title('The Elbow Method'); plt.xlabel('No. of Clusters'); plt.ylabel('Clustering Score (inertia)')
plt.show()
```

## 5. Final K-Means model (5 clusters)
Unsupervised: **train_test_split nei, y nei, accuracy nei** — inertia diye judge kori.

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans.fit(X)
pred = kmeans.predict(X)
pred
```

```python
df['Cluster'] = pred
print('Number of data points in each cluster:')
print(df['Cluster'].value_counts())
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
print('Inertia:', kmeans.inertia_)   # jato kom tato bhalo fit
```
