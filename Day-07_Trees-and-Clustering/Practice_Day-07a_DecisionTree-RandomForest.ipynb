## 1. Import + data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
df.head()
```

```python
df.columns
```

```python
sns.pairplot(df, hue='Kyphosis', palette='Set1')
```

## 2. X / y + target ke number banano
`Kyphosis` = target (absent→0, present→1). Baki 3 column = feature.

```python
X = df.drop('Kyphosis', axis=1)
y = df['Kyphosis']
y_numeric = y.map({'absent': 0, 'present': 1})
```

## 3. Split + Feature Scaling
`StandardScaler`: train-e `fit_transform`, test-e sudhu `transform` (data leakage edaj).
> Note: lab-e split-e random_state deoja chhilo na; ekhane `random_state=101` dilam jate result stable thake.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_numeric, test_size=0.30, random_state=101)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test  = sc.transform(X_test)
```

## 4. Single Decision Tree
`criterion='entropy'` -> information gain diye split quality maape.

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)
dtree.fit(X_train, y_train)
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

## 5. Tree ta dekho (visualize)

```python
from sklearn import tree
plt.figure(figsize=(16, 10))
tree.plot_tree(dtree, feature_names=list(X.columns), class_names=['absent','present'],
               rounded=True, filled=True)
plt.show()
```

## 6. Random Forest (onek tree = ensemble)
`n_estimators=100` -> 100 ta tree; majority vote. Single tree overfit kore, forest stable/bhalo hoy.

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)

print(confusion_matrix(y_test, rfc_pred))
print(classification_report(y_test, rfc_pred))
```

🧠 **KEY:** Random Forest-er accuracy usually single Decision Tree theke **beshi** (overfitting kom). Ei duita-r accuracy compare kore dekho.
